## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
is_5_digit_input = False
while (not is_5_digit_input):
    number_str = input("Введите 5-тизначное число")
    try:
        int(number_str)
    except:
        print("Введите число")
        continue
    if len(number_str) != 5: 
        print("Введите 5-тизначночное число")
        continue
    else:
        is_5_digit_input = True
        new_str = number_str[0] + number_str[3:0:-1] + number_str[len(number_str) -1]
        print(new_str)
        
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
is_positive_number_input = False
while (not is_positive_number_input):
    try:
        days = int(input("Введите число дней до отпуска"))
    except:
        print("Введите неотрицательное целое число")
        continue
    if days < 0:
        print("Введите неотрицательное целое число")
        continue
    else:
        is_positive_number_input = True
        holidays = 2 * (days // 7)
        if days % 7 == 6:
            holidays += 1
        print(f"До отпуска осталось {holidays} выходных")
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
# Чтобы можно было разломить плитку за один раз один из множителей должен ровняться одному измерению плитки, а второй множитель должен быть меньше либо равен второму измерению плитки
import math

is_possible = False
is_input_correct = False
input_str = input("Введите через запятую длину плитки, ширину плитки, а также размер желаемого куска")
while (not is_input_correct):
    try:
        input_str.replace(" ", "")
        input_array = input_str.split(",")
        if len(input_array) != 3:
            print("Введите через запятую 3 числа: длину плитки, ширину плитки, а также размер желаемого куска")
            continue
        dim_first = int(input_array[0])
        dim_second = int(input_array[1])
        area = int(input_array[2])
        is_input_correct = True
    except:
        print("Введите через запятую 3 числа: длину плитки, ширину плитки, а также размер желаемого куска")
        continue
    if dim_first * dim_second < area:
        print("False")
    else:
        square_root = int(math.sqrt(area))
        factor_first = 1
        factor_second = area
        while (factor_first <= square_root):
            if (factor_first == dim_first and factor_second <= dim_second 
                or factor_second == dim_first and factor_first <= dim_second 
                or factor_first == dim_first and factor_second < dim_first
                or factor_second == dim_second and factor_first < dim_first):
                is_possible = True
                break
            is_integer_factors_found = False
            while (not is_integer_factors_found and factor_first <= square_root):
                factor_first += 1
                if (area % factor_first == 0):
                    factor_second = area / factor_first
                    is_integer_factors_found = True
        print(is_possible)
                
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
arabic_to_rome = [(100, "C"), (50, "L"), (10, "X"), (5, "V"), (1, "I")]
print('enter positive arabic number less than 400')
is_input_correct = False
while (not is_input_correct):
  arabic = input()
  try:
    int_arabic = int(arabic)
  except:
    print('not a number')
    continue;
  if (int_arabic <=0 or int_arabic >= 400):
    print('not a positive number less than 400')
    continue;
  is_input_correct=True
  result = ''
  for index, value in enumerate(arabic_to_rome):
    factor = arabic_to_rome[index][0]
    rome_str = arabic_to_rome[index][1]
    quantity = int_arabic // factor
    int_arabic = int_arabic % factor
    if (quantity <= 3):
      result += rome_str * quantity
    else:
      result += rome_str + arabic_to_rome[index-1][1]
  print(result)
```

## Задание 5: Проверка на вещественное число
